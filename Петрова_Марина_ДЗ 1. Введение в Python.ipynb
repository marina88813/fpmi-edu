### Задача 4 (1). Проверка палиндрома

Напишите проверку, является ли строка палиндромом. Палиндром — это слово или фраза, которые одинаково читаются слева направо и справа налево.

#### Пример

**Ввод:**

```

```

```

```

````python
string = input('Введите строку для проверки: ')
````

````python
length_word = 0
for _ in string:
    length_word += 1
is_pal = True
for i in range(length_word // 2):
    if string[i] != string[length_word - 1 - i]:
        is_pal = False
        break
print("Палиндром" if is_pal else "Не палиндром")
````

### Задача 6 (0.5). Расширение файла

Вы получаете на вход имя файла. Необходимо вывести его расширение. Если расширения у файла нет, выводите `"Нет расширения"`.

Под расширением файла понимают определённую последовательность букв или цифр, отделенную от основного имени файла точкой.

**Примечание:**
- для файла `not_a_virus.jpg.exe` расширением является `.exe`, а не `.jpg.exe`.

#### Пример

**Ввод:**

```
absolutely_not_a_virus.exe
```

**Вывод:**

```
.exe
```

````python
filename = input("Введите имя файла: ")
````

````python
length_filename = 0
for _ in filename:
    length_filename += 1
last_dot = -1
research_last_dot = length_filename - 1
while research_last_dot >= 0:
    if filename[research_last_dot] == '.':
        last_dot = research_last_dot
        break
    research_last_dot -= 1
if last_dot != -1 and last_dot != length_filename - 1:
    extension = ''
    i = last_dot
    while i < length_filename:
        extension += filename[i]
        i += 1
    print(extension)
````

1.Множественное присваивание (multiple assignment)
Позволяет присвоить значения нескольким переменным в одной строке.
2.Распаковка кортежа (tuple unpacking)
Правая часть d, a, b, c автоматически создаёт кортеж, который затем распаковывается в переменные слева.
3.Вычисление правой части до присваивания (evaluation order)
Все значения в правой части вычисляются до того, как начинается присваивание переменным слева. Это гарантирует, что старые значения не будут потеряны в процессе обмена._

**Подсказка:** сохранять цены за все дни нельзя, но может это и не нужно, ведь у нас есть формат последнего сообщения!

````python
price = float(input())
min_price = price
max_price = price
summ_price = price
number_of_days = 1

while True:
    message = input().strip()
    if message == "#":
        break
    
    if message == "БОЛЬШЕ200":
        new_price = price + 200
    elif message == "МЕНЬШЕ100":
        new_price = price - 100
    elif message == "БЕЗ_ИЗМЕНЕНИЙ":
        new_price = price
    else:
        continue
    
    if new_price <= 0:
        new_price = price
    
    price = new_price
    
    if price < min_price:
        min_price = price
    if price > max_price:
        max_price = price
    
    summ_price = summ_price + price
    number_of_days = number_of_days + 1

average_price = summ_price / number_of_days
current_price = 0.6 * min_price + 0.4 * max_price

if current_price < average_price / 2:
    print("КУПИТЬ_10", file=sys.stderr)
elif current_price < average_price / 1.5:
    print("КУПИТЬ_7", file=sys.stderr)
elif current_price > average_price * 1.5:
    print("ПРОДАТЬ_6", file=sys.stderr)
elif current_price > average_price * 1.2:
    print("ПРОДАТЬ_3", file=sys.stderr)
else:
    print("ДЕРЖАТЬ", file=sys.stderr)
````
