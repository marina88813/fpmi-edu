## MathComplex (3 балла)

В этом задании вам нужно реализовать класс `MathComplex` для комплексных чисел. Они должны быть идентичны привычным нам комплексным числам из математики, а не объектам класса `complex`. Например, число `0+1i` должно печататься как `i`, число `42+0i` как просто `42`.

Вспомнить базовые свойства и операции комплексных чисел можно [здесь](http://www.mathprofi.ru/kompleksnye_chisla_dlya_chainikov.html) и [здесь](http://mathportal.net/index.php/kompleksnye-chisla/dejstviya-s-kompleksnymi-chislami).

Последовательно выполните указанные пункты:

1. реализовать методы `__init__` и нахождение модуля `__abs__`. Атрибуты назовите `real` и `imag`
2. красивый вывод на печать (достаточно реализовать `__repr__`). Для обозначения мнимой части используйте символ `i`
3. перегрузить операторы:
    - сложения
    - умножения
    - вычитания
    - деления (`__truediv__`)
<!--     - возведения в степень (считать, что только натуральную) -->

```python
import math

class MathComplex:
    def __init__(self, real=0, imag=0):
        self.real = real
        self.imag = imag

    def __abs__(self):
        return (self.real ** 2 + self.imag ** 2) ** 0.5

    def __repr__(self):
        if self.imag == 0:
            return f"{self.real}"

        i_str = "i" if self.imag == 1 else ("-i" if self.imag == -1 else f"{self.imag}i")

        if self.real == 0:
            return i_str

        sign = "+" if self.imag > 0 else ""
        return f"{self.real}{sign}{i_str}"

    def __add__(self, other):
        return MathComplex(self.real + other.real, self.imag + other.imag)

    def __sub__(self, other):
        return MathComplex(self.real - other.real, self.imag - other.imag)

    def __mul__(self, other):
        real_part = self.real * other.real - self.imag * other.imag
        imag_part = self.real * other.imag + self.imag * other.real
        return MathComplex(real_part, imag_part)

    def __truediv__(self, other):
        denom = other.real ** 2 + other.imag ** 2
        if denom == 0:
            raise ZeroDivisionError("Division by zero in complex numbers")
        real_part = (self.real * other.real + self.imag * other.imag) / denom
        imag_part = (self.imag * other.real - self.real * other.imag) / denom
        return MathComplex(real_part, imag_part)

    def __eq__(self, other):
        if not isinstance(other, MathComplex):
            return False
        return math.isclose(self.real, other.real) and math.isclose(self.imag, other.imag)
```

4. Вам нужно отнаследовать класс `ExtendedComplex` и реализовать в нем соответствующие специальные методы, чтобы арифметические операции из пункта 2 выполнялись еще и с операндами типа `int/float`.
```python
class ExtendedComplex(MathComplex):

    def __radd__(self, other):
        res = self + other
        return res

    def __rsub__(self, other):
        if isinstance(other, (int, float)):
            res = ExtendedComplex(other - self.real, -self.imag)
            print(f"[Правое вычитание] ({other}) - ({self}) = {res}")
            return res
        res = super().__sub__(other)
        return res

    def __rmul__(self, other):
        res = self * other
        return res

    def __rtruediv__(self, other):
        if isinstance(other, (int, float)):
            other_complex = ExtendedComplex(other, 0)
            res = other_complex / self
            print(f"[Правое деление] ({other}) / ({self}) = {res}")
            return res
        res = super().__truediv__(other)
        return res

    def __add__(self, other):
        if isinstance(other, (int, float)):
            res = ExtendedComplex(self.real + other, self.imag)
            print(f"[Сложение с числом] ({self}) + ({other}) = {res}")
            return res
        res = super().__add__(other)
        return ExtendedComplex(res.real, res.imag)

    def __sub__(self, other):
        if isinstance(other, (int, float)):
            res = ExtendedComplex(self.real - other, self.imag)
            print(f"[Вычитание числа] ({self}) - ({other}) = {res}")
            return res
        res = super().__sub__(other)
        return ExtendedComplex(res.real, res.imag)

    def __mul__(self, other):
        if isinstance(other, (int, float)):
            res = ExtendedComplex(self.real * other, self.imag * other)
            print(f"[Умножение на число] ({self}) * ({other}) = {res}")
            return res
        res = super().__mul__(other)
        return ExtendedComplex(res.real, res.imag)

    def __truediv__(self, other):
        if isinstance(other, (int, float)):
            r = self.real / other
            i = self.imag / other
            r = int(r) if r % 1 == 0 else r
            i = int(i) if i % 1 == 0 else i
            res = ExtendedComplex(r, i)
            print(f"[Деление на число] ({self}) / ({other}) = {res}")
            return res

        res = super().__truediv__(other)
        r = int(res.real) if res.real % 1 == 0 else res.real
        i = int(res.imag) if res.imag % 1 == 0 else res.imag
        return ExtendedComplex(r, i)

    def _coerce(self, other):
        if isinstance(other, (int, float)):
            return other, 0
        elif isinstance(other, MathComplex):
            return other.real, other.imag
        return None, None

    def __eq__(self, other):
        o_real, o_imag = self._coerce(other)
        if o_real is None:
            res = False
        else:
            res = (self.real == o_real and self.imag == o_imag)
        print(f"[Сравнение ==] ({self}) == ({other}) -> {res}")
        return res

    def __gt__(self, other):
        o_real, o_imag = self._coerce(other)
        if o_real is None:
            res = "результат не определен"
        else:
            cond1 = (self.real > o_real and self.imag >= o_imag)
            cond2 = (self.imag > o_imag and self.real >= o_real)

            if cond1 or cond2:
                res = True
            elif self.real == o_real and self.imag == o_imag:
                res = False
            else:
                res = "результат не определен"

        print(f"[Сравнение >] ({self}) > ({other}) -> {res}")
        return res

    def __lt__(self, other):
        o_real, o_imag = self._coerce(other)
        if o_real is None:
            res = "результат не определен"
        else:
            cond1 = (self.real < o_real and self.imag <= o_imag)
            cond2 = (self.imag < o_imag and self.real <= o_real)

            if cond1 or cond2:
                res = True
            elif self.real == o_real and self.imag == o_imag:
                res = False
            else:
                res = "результат не определен"

        print(f"[Сравнение <] ({self}) < ({other}) -> {res}")
        return res

    def __ge__(self, other):
        if self == other:
            res = True
        else:
            res = self > other
        print(f"[Сравнение >=] ({self}) >= ({other}) -> {res}")
        return res

    def __le__(self, other):
        if self == other:
            res = True
        else:
            res = self < other
        print(f"[Сравнение <=] ({self}) <= ({other}) -> {res}")
        return res
```

## MyRange (5 баллов)

6. Реализуйте класс `Range` и тесты к нему, который будет аналогом обычного `range`. Подробности в ячейке ниже: ориентируйтесь на перечисленные магические методы, которые нужно реализовать. Вам понадобится также сделать `RangeIterator`, экземпляры которого должны возвращаться методами `__iter__` и `__reversed__`. Проверка вхождения не должна вызывать итератор!

**Указание:** реализация возможности брать срез **целиком аналогично** обычному **range** требует дополнительных пересчетов индексов! В примерах ниже такие случаи специально опущены - поищите их самостоятельно :) Вспомните, что возвращает взятие среза в различных случаях.

```python
class RangeIterator:
    def __init__(self, start, stop, step, is_reversed=False):
        if step > 0 and start < stop:
            length = (stop - start - 1) // step + 1
        elif step < 0 and start > stop:
            length = (start - stop - 1) // (-step) + 1
        else:
            length = 0

        self.count = length

        if not is_reversed:
            self.current = start
            self.step = step
        else:
            self.current = start + (length - 1) * step if length > 0 else start
            self.step = -step

    def __next__(self):
        if self.count <= 0:
            raise StopIteration

        res = self.current
        self.current += self.step
        self.count -= 1
        return res

    def __iter__(self):
        return self


class Range:
    def __init__(self, *args):
        if len(args) == 1:
            self.start = 0
            self.stop = args[0]
            self.step = 1
        elif len(args) == 2:
            self.start = args[0]
            self.stop = args[1]
            self.step = 1
        elif len(args) == 3:
            self.start = args[0]
            self.stop = args[1]
            self.step = args[2]
            if self.step == 0:
                raise ValueError("range() arg 3 must not be zero")
        else:
            raise TypeError(f"range expected at most 3 arguments, got {len(args)}")

        if self.step > 0 and self.start < self.stop:
            self._len = (self.stop - self.start - 1) // self.step + 1
        elif self.step < 0 and self.start > self.stop:
            self._len = (self.start - self.stop - 1) // (-self.step) + 1
        else:
            self._len = 0

    def __len__(self):
        return self._len

    def __repr__(self):
        if self.step == 1:
            return f"range({self.start}, {self.stop})"
        return f"range({self.start}, {self.stop}, {self.step})"

    def __iter__(self):
        return RangeIterator(self.start, self.stop, self.step, is_reversed=False)

    def __reversed__(self):
        return RangeIterator(self.start, self.stop, self.step, is_reversed=True)

    def __contains__(self, value):
        if not isinstance(value, int):
            return False

        if self.step > 0 and not (self.start <= value < self.stop):
            return False
        if self.step < 0 and not (self.stop < value <= self.start):
            return False

        return (value - self.start) % self.step == 0

    def __eq__(self, other):
        if not isinstance(other, Range):
            return False
        if len(self) == 0 and len(other) == 0:
            return True
        return len(self) == len(other) and self.start == other.start and self.step == other.step

    def __getitem__(self, index):
        if isinstance(index, slice):
            start, stop, step = index.indices(len(self))

            if len(self) == 0 or (start >= stop and step > 0) or (start <= stop and step < 0):
                return Range(0, 0, 1)

            new_start = self.start + start * self.step
            new_stop = self.start + stop * self.step
            new_step = step * self.step

            return Range(new_start, new_stop, new_step)

        elif isinstance(index, int):
            if index < 0:
                index += len(self)
            if index < 0 or index >= len(self):
                raise IndexError("range object index out of range")

            return self.start + index * self.step
        else:
            raise TypeError(f"sequence indices must be integers or slices, not {type(index).__name__}")
```

## Дополнительная задача повышенной трудности

Напишите класс `HardRange` таким образом, чтобы его работа была целиком аналогична настоящему `range` с точки зрения использования срезов. В качестве примера работы `range`:

```python
range(55, -31, -7)[-14:2:-3]
```

```python
class HardRange:
    def __init__(self, *args):
        if len(args) == 1:
            start, stop, step = 0, args[0], 1
        elif len(args) == 2:
            start, stop, step = args[0], args[1], 1
        elif len(args) == 3:
            start, stop, step = args[0], args[1], args[2]
            if step == 0:
                raise ValueError("HardRange() arg 3 must not be zero")
        else:
            raise TypeError(f"HardRange expected at most 3 arguments, got {len(args)}")

        self.start = start
        self.stop = stop
        self.step = step
        self._elements = []

        curr = self.start
        if self.step > 0:
            while curr < self.stop:
                self._elements.append(curr)
                curr += self.step
        else:
            while curr > self.stop:
                self._elements.append(curr)
                curr += self.step

    def __len__(self):
        return len(self._elements)

    def __repr__(self):
        if self.step == 1:
            return f"HardRange({self.start}, {self.stop})"
        return f"HardRange({self.start}, {self.stop}, {self.step})"

    def __getitem__(self, index):
        if isinstance(index, slice):
            sliced_list = self._elements[index]

            slice_step = index.step if index.step is not None else 1

            if not sliced_list:
                return HardRange(0, 0, self.step * slice_step)

            if len(sliced_list) > 1:
                new_step = sliced_list[1] - sliced_list[0]
            else:
                new_step = self.step * slice_step

            new_start = sliced_list[0]
            new_stop = sliced_list[-1] + new_step

            return HardRange(new_start, new_stop, new_step)

        elif isinstance(index, int):
            return self._elements[index]
        else:
            raise TypeError(f"Indices must be integers or slices, not {type(index).__name__}")

    def __eq__(self, other):
        if not isinstance(other, HardRange):
            return False
        return self._elements == other._elements
```
